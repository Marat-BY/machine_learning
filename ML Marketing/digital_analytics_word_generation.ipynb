## Search Engine Marketing (SEM) and Campaigns
### What is SEM?
SEM is one of the most effective ways to grow your business in an increasingly competitive marketplace. It is the process of gaining website traffic by purchasing ads on search engines. Google AdWords is one of the most popular forms of online advertising (you bid for ad placement in a search engine's sponsored links for keywords related to your business, then you pay to the search engine fee for each click or other type of action fee. It can be click, views and etc.).

Paying per click in part of the situations is very expensive. And so it is very important to plan and group your keywords and ads in such way that you control you cost and have a profitable campaigns.

Remeber: ** the more relevant your ads are the lower your cost per click **


### Setting up a Search Campaign.
Before setting up a campaign you will typically come up with a budget, pick keywords and check out the competition. Also you will make sure that your landing page is in order. Then you can set up your campaign: you will write your first ad, fix it details and then set up conversion tracking. This work cover the process of generating a large number of keywords and ads, based on the different landing page you have. 


#### Ad Groups
To get started with a search campaign you first need to develop a plan for your ad groups. This is the basic unit of organization where you map keywords, ads and landing pages. But what are these three components exactly?

1. Keywords: the user's intention, or otherwords what they are looking for?
2. Ads: your promise to the user for that intention.
3. Landing Pages: your delivery of the promise.

Mapping those three elements properly is around 70% - 80% of the work process. If you get this right then maintenance and changes become much more easier. This means that, in order to be specific and relevant users, you need to promote the product you have and mainly focus on sending the right person to the right page, through the message or expectetion. A good campaign structure reflects and follows the structure of the website which is essentially reflecting the business strategy. Besides the fact, note that every ad group needs to be part of a campaign. At the campaign level, you have settings that govern all ad groups in the campaign such as language targeting, location targeting, devices, timezone and other valuable criteria.

# 2. Implementation:

The first step is to get a list of words that user or users will use to express their desire in buying a furniture (sofas, beds, tables or other types of furniture)

```python
# List of words to pair with products

words = ['buy furniture', 'discounts on', 'discounts furniture New York', 'Beds', 'Low cost', 'price',
         'promotion', 'furniture shopping', 'furniture shop', 'promo']

print(words)
```

Creating all possible combinations of keywords for adwertising campaigns is hard and stressfull. But not with Python. No we know how to translate campaign briefs into Python data structures and can imagine the resulting DataFrames that we need to create.

Now that we have 'brainstormed' the words that work well with the brief that we recieved, it is now time to combine them with the product names to generate meaningfull search keywords. We want to combineevery word with every product once before and once after as you can see in the exmaple above.

As a quick reminder for the product 'recliners' and the words 'buy' and 'price' for example we would need to generate the following combinations:

* Buy sofas
* sofas buy
* price furniture
* furniture price 
* And etc for all of the words and products the company possibly have or might have

```python
#list of products
products = ['sofas', 'convertible sofas', 'love seats', 'recliners', 'sofa beds', 'tables', 'computer tables', 'chairs', 'office chairs']

# Empty list
keywords_list = []

# Looping through products
for product in products:
    # Looping throug words
    for word in words:
        # Create and append different kinds of combinations
        keywords_list.append([product, product + ' ' + word])
        keywords_list.append([product, word + ' ' + product])
```

```python
# Inspect keyword list
from pprint import pprint
pprint(keywords_list)
```

Let's conver the list of lists of created key words into DataFrame object.
Now we want to convert this list of list into DataFrame so we can easily manipulate it and manage the final output of the generated words.

```python
# To convert the generated words we need special classmethod DataFrame.from_records()
import pandas as pd

# Create a DataFrame from list
keywords_df = pd.DataFrame.from_records(keywords_list)

# Explore the DataFrame
keywords_df.head()
```

Rename the created columns of the DataFrame. 
Before we can upload this table of keywords, we will need to give the columns meaningful names. If we inspect the DataFrame we just created above, we can see that the columns are currently named 0 and 1. Ad Group (e.g. 'Sofas') and Keyword (e.g. 'sofas buy furniture') are much more approprite names.

```python
# Rename the columns of the DataFrame
keywords_df.rename(columns = {0 : 'Ad Group', 1 : 'Keyword'}, inplace = True)

# Investigate the result of renaming
keywords_df.head()
```

Add a campaign column. So now we need to add some additional information to our DataFrame. We need a new column called Campaign for the campaign name. We want campaign names to be descriptive of our group of keywords and products, so let's call this campaign 'SEM_Sofas'.

```python
# Adding a campaign column
keywords_df['Campaign'] = 'SEM_Sofas'
```

```python
keywords_df.head()
```

### Create the match type column

There are different keyword match types. One is exact match, which is for matching the exact term or are close variations of that exact term. Another match type is broad match, which means ads may show on searches that include misspellings, synonyms, related searches nad other types of relevant variations.

[Google Adwords Documentation on match types of words.](https://support.google.com/google-ads/answer/2497836?hl=en)


In general, the broader the match type the more traffic potential that keyword will have, since your ads may be triggered more often. Conversely, a narrower match type means that your ads may show less often - but when they do, they are likely to be more related to someone's search.


Since the client is tight on budget we want to mek sure all the keywords are in exact match at the beginning.

```python
# Add a criterion type column
keywords_df['Criterion Type'] = 'Exact'
```

### Duplicate all the keywords into 'phrase' match

The great thing about exact match is that it is very specific, and we can control the process very well. The tradeoff, however is that:

1. The search volume for exact match is lower than other match types
2. We can not possibly think of all the ways in which people search and so we are probably missing out some high-quality keywords.

So it is good to use another match called *phrase match* as a discovery mechanism to allow our ads to be triggered by keywords that include our exact match keywords, together with anything before (or after) them.

Moreover you can explore with modified broad match, broad match, and negative match types for better visibility and control of your campaigns.

```python
# Copy of keywords DataFrame
keywords_phrase = keywords_df.copy()

# Changing criterion type match to phrase
keywords_phrase['Criterion Type'] = 'Phrase'
```

```python
keywords_df_final = keywords_df.append(keywords_phrase, ignore_index = False)
```

```python
keywords_df_final.head()
```

### Final phase and summarizing

To upload created campaigns we need to save it as a CSV file. Then we will be able to import it to Adwords editor, Yandex editor or BingAds editor. There is also the option of pasting the data into the editor if we want, but having easy access to the saved data is great so let's save to CSV file.

Looking at a summary of our campaign structure is good now that we have wrapped up our keyword work. We can do that by grouping by ad group and criterion type and counting by keyword. This summary shows us that we assigned a specific keywords to specific ad grops which are part of a campaign. In essence we are telling Google that we want any of the words in each ad group to trigger one of the ads in the same ad group. Separately, we will have to create another table for ads, which I will cover in Part 2. 

```python
# Saving the file
keywords_df_final.to_csv('keywords_for_campaigns.csv', index = False)

# Summary of campaign creation work
summary = keywords_df_final.groupby(['Ad Group', 'Criterion Type'])['Keyword'].count()
print(summary)
```
